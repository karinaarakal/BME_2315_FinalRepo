## Data Analyis: 

### Methods
The machine learning technique we are using is Principal Component Analysis (PCA). PCA takes high-dimensional gene expression data and decomposes it into new features that find the directions of greatest variance in the dataset. In our case, PCA helps visualize patterns in VEGF/MAP-Kinase pathway activity across different cancers relative to lung cancer. This lets us see how similar or different the tumors are based on angiogenesis-related gene expression.

*What is this method optimizing? How does the model decide it is "good enough"?*
PCA maximizes variance while minimizing the projection error. It’s considered “good enough” when the first principal components explain a meaningful amount of the total variance and the plot shows clear structure or separation between groups.


### Analysis
*(Describe how you analyzed the data. This is where you should intersperse your Python code so that anyone reading this can run your code to perform the analysis that you did, generate your figures, etc.)*

To analyze the data, we used the TCGA dataset provided in class. First, we loaded the gene expression file and the metadata file, which includes cancer type information. We then selected several cancer types (LUAD, LIHC, KIRC, HNSC, COAD, STAD, SKCM) so we could compare lung cancer to cancers from different parts of the body. Based on the angiogenesis review paper, we focused on genes connected to VEGF signaling and inflammation, since both can promote angiogenesis in tumors. The genes available in the dataset that matched this pathway were VEGFA, KDR, NRP1, NRP2, EGFR, IL6, IL8, IL1B, IL1R1, and ANGPT1. We filtered the dataset to include only these genes and the samples from the cancer types of interest. Then, we combined the gene data with the metadata so each sample had both its gene values and its cancer type label. To explore the data, we created boxplots to compare expression levels of each gene across cancer types. This helps show how VEGF-related and inflammation-related genes vary between different cancers. Next, we used PCA to reduce the gene data into two components and then plotted PC1 vs PC2 to look for patterns across cancer types. We scaled the gene values before PCA so they were comparable. We also printed the amount of variation explained by PC1 and PC2.


```python
# Exploratory data analysis (EDA) on a cancer dataset
# Loading the files and exploring the data with pandas
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import pandas as pd
import seaborn as sns
# Load the data
data = pd.read_csv('/Users/karinaarakal/Downloads/GSE62944_subsample_topVar_log2TPM.csv', index_col=0, header=0)  # can also use larger dataset with more genes
metadata_df = pd.read_csv('/Users/karinaarakal/Downloads/GSE62944_metadata.csv', index_col=0, header=0)
print(data.head())

# Explore the data
print(data.shape)
print(data.info())
print(data.describe())

# Explore the metadata
print(metadata_df.info())
print(metadata_df.describe())

# Subset the data for specific cancer types
cancer_types = ['LIHC', 'KIRC']

# From metadata, keep only samples whose cancer_type is in our list
cancer_samples = metadata_df[metadata_df['cancer_type'].isin(cancer_types)].index
print(f"Selected samples: {len(cancer_samples)}")

# Subset the main data to include only these samples
BRCA_data = data[cancer_samples]

# Subset by index
desired_gene_list = [
    'VEGFA','KDR',         # VEGF axis
    'NRP1','NRP2',         # co-receptors
    'EGFR',                # MAPK linkage
    'IL6','IL8','IL1B','IL1R1',  # interleukins
    'ANGPT1'               # Ang-Tie pathway
]
gene_list = [gene for gene in desired_gene_list if gene in BRCA_data.index]

for gene in desired_gene_list:
    if gene not in gene_list:
        print(f"Warning: {gene} not found in the dataset.")

# .loc[] is the method to subset by index labels
BRCA_gene_data = BRCA_data.loc[gene_list]
print(BRCA_gene_data.head())

# Merge with metadata
cancer_samples = data.columns.intersection(cancer_samples)
BRCA_data = data[cancer_samples]
BRCA_metadata = metadata_df.loc[cancer_samples]
BRCA_merged = BRCA_gene_data.T.merge(BRCA_metadata, left_index=True, right_index=True)
print(BRCA_merged.head())

# Boxplots
#for g in gene_list:
  #  sns.boxplot(data=BRCA_merged, x="cancer_type", y=g)
  #  plt.title(f"{g} Expression Across Cancer Types")
  #  plt.show()
    
# PCA
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

X = BRCA_merged[gene_list].fillna(BRCA_merged[gene_list].median())
Xz = StandardScaler().fit_transform(X)

pca = PCA(n_components=2)
pcs = pca.fit_transform(Xz)

pca_df = pd.DataFrame(pcs, columns=["PC1","PC2"], index=BRCA_merged.index)
pca_df["cancer_type"] = BRCA_merged["cancer_type"]

sns.scatterplot(data=pca_df, x="PC1", y="PC2", hue="cancer_type", s=70, alpha=0.85)
plt.title("PCA of selected angiogenesis/MAPK/IL genes")
plt.tight_layout()
plt.show()

print("Explained variance (PC1, PC2):", pca.explained_variance_ratio_)

#silhouette

pcs = pca_df[["PC1", "PC2"]].values
X = pcs
n_clusters = 2

# -------------------------------------------------
# Run KMeans
# -------------------------------------------------
kmeans = KMeans(n_clusters=n_clusters, random_state=10)
cluster_labels = kmeans.fit_predict(X)

# Silhouette scores
silhouette_avg = silhouette_score(X, cluster_labels)
print(f"Average silhouette score for 2 clusters: {silhouette_avg:.4f}")
sample_silhouette_values = silhouette_samples(X, cluster_labels)


# -------------------------------------------------
# PLOTTING
# -------------------------------------------------
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 7))

# ==============================
# 1. Silhouette Plot
# ==============================
ax1.set_xlim([-0.2, 1])
ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

y_lower = 10
for i in range(n_clusters):
    cluster_sil_vals = sample_silhouette_values[cluster_labels == i]
    cluster_sil_vals.sort()

    size_i = cluster_sil_vals.shape[0]
    y_upper = y_lower + size_i

    color = cm.nipy_spectral(float(i) / n_clusters)
    ax1.fill_betweenx(
        np.arange(y_lower, y_upper),
        0,
        cluster_sil_vals,
        facecolor=color,
        edgecolor=color,
        alpha=0.7
    )

    ax1.text(-0.1, y_lower + 0.5 * size_i, str(i))
    y_lower = y_upper + 10

ax1.axvline(x=silhouette_avg, color="red", linestyle="--")
ax1.set_title("Silhouette Plot (k = 2)")
ax1.set_xlabel("Silhouette coefficient (s = 0.5746)")
ax1.set_ylabel("Cluster label")
ax1.set_yticks([])

# ==============================
# 2. Scatter Plot in PCA Space
# ==============================
colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)

ax2.scatter(
    X[:, 0], X[:, 1],
    c=colors,
    edgecolor="k",
    s=60,
    alpha=0.9
)

# Add cluster centers
centers = kmeans.cluster_centers_
ax2.scatter(
    centers[:, 0], centers[:, 1],
    marker="X",
    c="white",
    s=300,
    edgecolor="k"
)
for idx, c in enumerate(centers):
    ax2.text(c[0], c[1], str(idx), fontsize=12,
             ha="center", va="center", color="black")

ax2.set_title("KMeans Clusters in PCA Space")
ax2.set_xlabel("PC1")
ax2.set_ylabel("PC2")

# =========================================
# 1. Extract PCA Loadings for Each Gene
# =========================================

loadings = pd.DataFrame(
    pca.components_.T,       # transpose so each row = gene
    index=gene_list,         # gene names
    columns=["PC1", "PC2"]   # name the components
)

print("\n=== PCA LOADINGS ===")
print(loadings)

# =========================================
# 2. Rank Genes by Contribution to PC1 & PC2
# =========================================

loadings["abs_PC1"] = loadings["PC1"].abs()

top_PC1 = loadings.sort_values("abs_PC1", ascending=False)

print("\n=== Top Contributing Genes to PC1 ===")
print(top_PC1.head())



# =========================================
# Barplot of Top Contributing Genes for PC1
# =========================================

top_n = 10  # choose how many genes to show
sorted_PC1 = loadings["PC1"].sort_values(key=np.abs, ascending=False)[:top_n]

plt.figure(figsize=(8,6))
sorted_PC1.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Top Contributing Genes to PC1")
plt.ylabel("Loading Score")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
